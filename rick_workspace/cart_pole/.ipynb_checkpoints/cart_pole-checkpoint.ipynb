# Cart-Pole Demo
"Reinforcement learning (RL) is the subfield of machine learning concerned with decision making and motor control. It studies how an agent can learn how to achieve goals in a complex, uncertain environment" - from  https://gym.openai.com/docs/

The cart pole is a simulation program. It demonstrates an agent takes random actions in an environment to learn how to keep the pole balanced in a moving cart.

```python
import gym
# Make cart pole environment
env = gym.make('CartPole-v0')


print(env.action_space) # Two valid actions: either 0 or 1 - applying force to the left or right
print(env.observation_space) # 4-dimensional boxâ€™s bounds for valid observations
print(env.observation_space.high) # The high values of cart_position, cart_velocity, pole_angle and pole_velocity
print(env.observation_space.low)

del env
```

```python
import gym

def main():

    # Make cart pole environment
    env = gym.make('CartPole-v0')
    
    # Run 20 episodes
    for episode in range(20):
        observation = env.reset()
    
        for t in range(100): 
            env.render()
            action = env.action_space.sample()# take a random action
            observation, reward, done, info = env.step(action)
            print(observation)
            if done:
                print("Episode finished after {} timesteps".format(t+1))
                break

if __name__ == "__main__":
    main()
```
